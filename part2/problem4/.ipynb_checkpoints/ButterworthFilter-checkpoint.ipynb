# Butterworth Filter
**First, create an array with a length of 1000 of normal random noise. A Butterworth bandpass filter of order 3 that filters frequencies outside the range of 300 to 3000. Apply discrete Fourier transform and display the result of applying this filter to the generated noise in the time/place range.**


The following code snippet import required third-party libraries:
+ `cv2` for reading and utilizing image processing functions
+ `numpy` for calculating and considering images as numerical matrixes
+ `matplotlib.pyplot` for showing images in the output
+ `signal.butter` for applying bandpass filter

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter
```

In the next few lines, we will produce a normal noise and display it.

```python
noise = np.random.normal(0, 1, 1000)

plt.plot(noise)
plt.show()
```

Next we will create a bandpass butterworth filter and apply to the noise.

```python
# Generate filter
b, a = butter(3, [300, 3000], 'bandpass', analog=True)
butter_filter = (np.concatenate((a,b)))
# Repeat filter
butter_filter = np.repeat(butter_filter, 100)

# Get fourier transform
filter_freq = np.fft.fft(butter_filter)

plt.plot(np.abs(filter_freq))
plt.title("Magnitude of filter")
plt.show()

nosise_freq = np.fft.fft(noise)

# Apply filter
result = noise_freq*filter_freq[:-100]

plt.plot(np.abs(np.fft.ifft(result)))
plt.title('Filtered Noise')
plt.show()
```
